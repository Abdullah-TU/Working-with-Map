# Opening images in a centain range of tiles

```python
# opening images in a centain range of tiles
import time

zoom=6    # zoom level
imgs =[]
for x in range(34,37):       # x coordinate in the range 34-36
    for y in range(21,24):    # y coordinates in the range 21-24  
        imgs.append(request.urlopen('https://maps.wikimedia.org/osm-intl/{}/{}/{}.png'.format(zoom, x,y)).read())
        time.sleep(1)
        
# display images one by one

from IPython.display import display

for data in imgs:
    img=Image.open(BytesIO(data))
    display(img)
```

```python
from PIL import Image
from io import BytesIO

mapimg = Image.new('RGB', (3*256,3*256))
mapimg.paste(Image.open(BytesIO(imgs[3])))
mapimg
```

```python
from PIL import Image
from io import BytesIO

mapimg = Image.new('RGB', (3*256,3*256))
mapimg.paste(Image.open(BytesIO(imgs[3])), (256,256))
mapimg
```

```python
mapimg= Image.new('RGB', (3*256, 3*256))
for col in range(3):
    for row in range(3):
        img= imgs[3*col+row]
        mapimg.paste(Image.open(BytesIO(img)),(col*256, row*256))
mapimg.save('fullarea.png')
mapimg
```
